## Opencv

```python
# Открытие изображения с помощью opencv
image = cv2.imread(image_file)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Уменьшение размера в два раза
resized_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
resized_image
```

```python
# Произвольное изменение размера
resized_image = cv2.resize(image, (300, 200))
resized_image
```

```python
# Преобразование в серое изображение
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
# Применение Гауссового размытия
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)
blurred_image
```

```python
# Получение размеров изображения
(height, width) = image.shape[:2]

# Определение центра поворота
center = (height // 2, width // 2)

# Установка угла поворота и масштаба
angle = 45  # например, 45 градусов
scale = 1.0

# Создание матрицы поворота
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# Применение матрицы к изображению
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
rotated_image
```

```python
# Определение координат для обрезки (x_start, y_start, width, height)
x_start, y_start, width, height = 10, 10, 130, 160  # Например, обрезка области 300x300 пикселей

# Обрезка изображения
cropped_image = image[y_start:y_start + height, x_start:x_start + width]
cropped_image
```

## Рисование рамки

Шаг 1: Импорт библиотек и загрузка изображения

```python
import cv2

# Загрузка изображения
image = cv2.imread('logo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Проверка, загрузилось ли изображение
if image is None:
    print("Ошибка: изображение не найдено.")
    exit()
```

Шаг 2: Определение координат рамки

```python
# Определение координат для рамки (x_start, y_start, width, height)
x_start = 20
y_start = 30
width = 110
height = 160

# Вычисление координат правого нижнего угла
x_end = x_start + width
y_end = y_start + height
```

Шаг 3: Рисование рамки

```python
# Рисование рамки на изображении (цвет в BGR, толщина линии)
color = (0, 255, 0)  # Зеленый цвет
thickness = 2

cv2.rectangle(image, (x_start, y_start), (x_end, y_end), color, thickness)
```

Шаг 4: Добавление текстовой надписи

```python
# Определение текста и его параметров
text = "Это рамка"
font = cv2.FONT_HERSHEY_COMPLEX
font_scale = 0.7
font_color = (255, 255, 255)  # Белый цвет
line_type = 1

# Позиция текста (можно изменить по желанию)
text_x = x_start + 10
text_y = y_start - 10

# Добавление текста на изображение
cv2.putText(image, text, (text_x, text_y), font, font_scale, font_color, line_type)
```

### Воспроизведение видео

```python
import cv2
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
import time

# Replace 'your_video.mp4' with the name of your uploaded video file
video_path = 'video.MP4'

# Open the video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Read and display frames
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if there are no more frames

    # Convert BGR (OpenCV format) to RGB (Matplotlib format)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Clear previous output
    clear_output(wait=True)

    # Display the current frame using cv2_imshow
    cv2_imshow(frame)

    # Pause to create a delay between frames (adjust for playback speed)
    time.sleep(0.03)  # Adjust this value for playback speed

# Release the video capture object
cap.release()
```
